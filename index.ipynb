## Variable Descriptions

This data frame contains the following columns:

#### crim  
per capita crime rate by town.

#### zn  
proportion of residential land zoned for lots over 25,000 sq.ft.

#### indus  
proportion of non-retail business acres per town.

#### chas  
Charles River dummy variable (= 1 if tract bounds river; 0 otherwise).

#### nox  
nitrogen oxides concentration (parts per 10 million).

#### rm  
average number of rooms per dwelling.

#### age  
proportion of owner-occupied units built prior to 1940.

#### dis  
weighted mean of distances to five Boston employment centers.

#### rad  
index of accessibility to radial highways.

#### tax  
full-value property-tax rate per $10,000.

#### ptratio  
pupil-teacher ratio by town.

#### black  
1000(Bk - 0.63)^2 where Bk is the proportion of blacks by town.

#### lstat  
lower status of the population (percent).

#### medv  
median value of owner-occupied homes in $10000s.

```python
comparecols = ['crim','ptratio','nox','dis','tax', 'lstat', 'black', 'age']

df_scatter = df.loc[:,comparecols]
pd.plotting.scatter_matrix(df_scatter, figsize=(15,15));
```

```python
comparecols = ['crim','ptratio']



df_scatter = df.loc[:,comparecols]
df_scatter.plot(x='ptratio',y='crim', kind='scatter', title='Pupil/Teacher Ratio vs Crime', figsize=(14,6));

```

```python
comparecols = ['dis','nox']


df_scatter = df.loc[:,comparecols]
df_scatter.plot(x='dis', y='nox', kind='scatter', figsize=(12,10),title='Distance from 5 Employment Centers versus Nitrogen Levels');
```

```python
comparecols = ['age','lstat']


df_scatter = df.loc[:,comparecols]
df_scatter.plot(x='age', y='lstat', kind='scatter', figsize=(12,10));
```

```python
#split up into charles and not charles data frames

df_charles = df[df['chas']==1]
df_not_charles = df[df['chas']==0]

a = df_charles.describe()
b = df_not_charles.describe()
print('Charles statistics minus Not Charles statistics')
a-b
```

```python
#split up into charles and not charles data frames

breakpoint = df.crim.median()

df_crim = df[df['crim']>=breakpoint]
df_not_crim = df[df['crim']<breakpoint]

a = df_crim.describe()
b = df_not_crim.describe()
print('Top 50% Crime statistics minus Bottom 50% crime statistics')
a-b
```

```python
x1 = df_crim.tax
x2 = df_not_crim.tax

y1 = df_crim.ptratio
y2 = df_not_crim.ptratio


# Define a new figure with matplotlib's .plot() function. Set the size of figure space
new_figure = plt.figure(figsize=(10,4))

# Add a subplot to the figure - a new axes
ax = new_figure.add_subplot(121)

# Add a second subplot to the figure - a new axes
ax2 = new_figure.add_subplot(122)

# Generate a line plot on first axes
ax.scatter(x1, y1, color='lightblue', marker='o')

# Draw a scatter plot on 2nd axes
ax2.scatter(x2,y2, color='red', marker='o')

# Set the limits of x and y for first axes
ax.set_xlim(0, 800), ax.set_ylim(0,35)

# Set the limits of x and y for 2nd axes
ax2.set_xlim(0, 800), ax2.set_ylim(0,35)

# Show the plot
plt.show()
```
